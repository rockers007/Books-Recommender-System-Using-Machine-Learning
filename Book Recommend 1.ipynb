```python
book_data=pd.read_csv("data/BX-Books.csv",sep=";",on_bad_lines='skip' , encoding='latin-1')
```

```python
user_data=pd.read_csv("data/BX-Users.csv",sep=";",on_bad_lines='skip' , encoding='latin-1')
```

```python
user_data.head()
```

```python
rating_data=pd.read_csv("data/BX-Book-Ratings.csv",sep=";",on_bad_lines='skip' , encoding='latin-1')
```

```python
rating_data
```

```python
rating_data['User-ID'].unique().shape
```

```python
x=rating_data['User-ID'].value_counts()>50
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
rating_data = rating_data[rating_data['User-ID'].isin(y)]
```

```python
rating_data
```

```python
pip install jupyter_contrib_nbextensions
```

```python
rating_book=rating_data.merge(book_data,on='ISBN')
```

```python
rating_book= rating_book[['User-ID','Book-Rating','ISBN','Book-Title','Image-URL-L']]
```

```python
rating_book
```

```python
rating_book.rename(columns={"Book-Title":'title',
                      'User-ID':'user_id',
                     "Book-Rating":'rating',
                     "Image-URL-L":"image_url"},inplace=True)
```

```python
rating_book
```

```python
number_rating = rating_book.groupby('title')['rating'].count().reset_index()
```

```python
number_rating
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating
```

```python
final_rating = rating_book.merge(number_rating, on='title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
```

```python
final_rating
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0, inplace=True)
```
